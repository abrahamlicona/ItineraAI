##TabNet (Pretrainer):
Es un modelo de deep learning diseñado especialmente para datos tabulares.
Aquí se usa en modo no supervisado (TabNetPretrainer) para aprender representaciones latentes ("embeddings") de los datos.
Aprende a reconstruir la entrada original desde una representación comprimida.
Esta representación captura las relaciones complejas entre variables, incluso categóricas codificadas, de forma eficiente.
##KMeans:
Una vez obtenidas las representaciones latentes, se aplica KMeans para hacer clustering sobre esas representaciones.
Se prueban valores de k de 3 a 5 clusters, evaluando cada uno con el Silhouette Score.
Se elige el número de clusters con mejor desempeño (mayor Silhouette Score).

```python
# Colab Cell 1: Install all ML deps
!pip install scikit-learn pandas numpy torch pytorch-tabnet

# Colab Cell 2: Train TabNet pretrainer, cluster with KMeans, serialize pipeline
import pickle
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from pytorch_tabnet.pretraining import TabNetPretrainer
import torch
from google.colab import files

# Seeds
np.random.seed(42)
torch.manual_seed(42)

# 1) Load & prep data
df = pd.read_csv("/content/Reservaciones_df_limpio.csv", low_memory=False)
num_vars = ['h_num_per','h_num_adu','h_num_men','h_num_noc','h_tot_hab','h_tfa_total']
cat_vars = ['ID_Tipo_Habitacion','ID_canal','ID_Pais_Origen','ID_Segmento_Comp','ID_Agencia']
df = df[(df[num_vars] >= 0).all(axis=1) & (df["h_num_noc"] < 100)]
X = df[num_vars + cat_vars].copy()

# 2) Fit LabelEncoders on cat_vars
encoders = {}
for col in cat_vars:
    enc = LabelEncoder().fit(X[col])
    X[col] = enc.transform(X[col])
    encoders[col] = enc

# 3) Prepare array for TabNet (float32)
X_array = X.values.astype(np.float32)

# 4) Unsupervised TabNet pretraining
unsup = TabNetPretrainer(
    optimizer_fn=torch.optim.Adam,
    optimizer_params=dict(lr=1e-3),
    seed=42, verbose=0
)
unsup.fit(
    X_train=X_array, eval_set=[X_array],
    max_epochs=100, patience=10,
    batch_size=256, virtual_batch_size=128
)

# 5) Extract embeddings
embeddings, _ = unsup.predict(X_array)

# 6) KMeans on embeddings, pick best K
best_score = -1
for k in range(4,6):
    km = KMeans(n_clusters=k, random_state=42).fit(embeddings)
    score = silhouette_score(embeddings, km.labels_)
    if score > best_score:
        best_score, best_k, best_labels, best_km = score, k, km.labels_, km

df["Cluster_TabNet"] = best_labels

print(f"✔️ Best K={best_k}, Silhouette={best_score:.4f}")

# 7) Assemble pipeline dict
pipeline = {
    "encoders": encoders,
    "tabnet":   unsup,
    "kmeans":   best_km,
    "num_vars": num_vars,
    "cat_vars": cat_vars
}

# 8) Serialize pipeline.pkl
with open("pipeline.pkl","wb") as f:
    pickle.dump(pipeline, f)

# 9) Download to local
files.download("pipeline.pkl")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from pytorch_tabnet.pretraining import TabNetPretrainer
from pytorch_tabnet.tab_model import TabNetClassifier
import torch

# Semillas
np.random.seed(42)
torch.manual_seed(42)

# ---------------------
# Cargar y preparar datos
# ---------------------
df = pd.read_csv("/content/Reservaciones_df_limpio.csv", low_memory=False)
num_vars = ['h_num_per', 'h_num_adu', 'h_num_men', 'h_num_noc', 'h_tot_hab', 'h_tfa_total']
cat_vars = ['ID_Tipo_Habitacion', 'ID_canal', 'ID_Pais_Origen', 'ID_Segmento_Comp', 'ID_Agencia']
df = df[(df[num_vars] >= 0).all(axis=1) & (df["h_num_noc"] < 100)]

X = df[num_vars + cat_vars].copy()
for col in cat_vars:
    X[col] = LabelEncoder().fit_transform(X[col])

# TabNet requiere datos como np.float32
X_array = X.values.astype(np.float32)

# ---------------------
# Entrenamiento no supervisado
# ---------------------
unsupervised_model = TabNetPretrainer(
    optimizer_fn=torch.optim.Adam,
    optimizer_params=dict(lr=1e-3),
    seed=42,
    verbose=0
)

unsupervised_model.fit(
    X_train=X_array,
    eval_set=[X_array],
    max_epochs=100,
    patience=10,
    batch_size=256,
    virtual_batch_size=128
)

# ---------------------
# Obtener representaciones latentes (embeddings)
# ---------------------
# TabNet devuelve (embeddings, reconstructed_input)
raw_output = unsupervised_model.predict(X_array)
embeddings = raw_output[0]  # El primero es el embedding


# ---------------------
# Clustering con KMeans sobre embeddings
# ---------------------
best_score = -1
for k in range(3, 6):
    kmeans = KMeans(n_clusters=k, random_state=42).fit(embeddings)
    score = silhouette_score(embeddings, kmeans.labels_)
    print(f"K={k}, Silhouette Score: {score:.4f}")
    if score > best_score:
        best_k = k
        best_labels = kmeans.labels_
        best_score = score

df["Cluster_TabNet"] = best_labels
print(f"\n✅ Mejor K: {best_k} con Silhouette Score: {best_score:.4f}")
```

```python
def resumen_clusters(df, cluster_col):
    return df.groupby(cluster_col).agg({
        'h_num_per': 'mean',
        'h_num_adu': 'mean',
        'h_num_men': 'mean',
        'h_num_noc': 'mean',
        'h_tot_hab': 'mean',
        'h_tfa_total': 'mean',
        'ID_Tipo_Habitacion': lambda x: x.value_counts().index[0],
        'ID_canal': lambda x: x.value_counts().index[0],
        'ID_Pais_Origen': lambda x: x.value_counts().index[0]
    }).round(2)

print("\n📋 Resumen de perfiles (TabNet + KMeans):")
resumen_clusters(df, "Cluster_TabNet")
```

```python


print("\n Observaciones por cluster (TabNet + KMeans):")
print(df["Cluster_TabNet"].value_counts().sort_index())
```

```python
# Aplicar KMeans con k=4 directamente
kmeans_k4 = KMeans(n_clusters=4, random_state=42).fit(embeddings)
df["Cluster_TabNet_k4"] = kmeans_k4.labels_

# Silhouette Score para k=4
score_k4 = silhouette_score(embeddings, kmeans_k4.labels_)
print(f"\n🎯 K=4 Silhouette Score (recalculado sin reentrenar): {score_k4:.4f}")
```

```python
# Aplicar KMeans con k=5 directamente
kmeans_k5 = KMeans(n_clusters=5, random_state=42).fit(embeddings)
df["Cluster_TabNet_k5"] = kmeans_k5.labels_

# Silhouette Score para k=5
score_k5 = silhouette_score(embeddings, kmeans_k5.labels_)
print(f"\n🎯 K=5 Silhouette Score (recalculado sin reentrenar): {score_k5:.4f}")
```
